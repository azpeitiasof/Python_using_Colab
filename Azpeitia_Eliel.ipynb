```python
import pandas as pd # nos permite cargar datos como archivos de excel (formatos .xlsx y .csv)
import matplotlib.pyplot as plt # Nos permite realizar gr√°ficos con python
import seaborn as sns
```

```python
dataset = pd.read_excel('Latam Christmas Gifts Dataset.xlsx')
```

```python
dataset.head(5)
```

```python
Regalos = dataset.groupby(['gift_name'])['age'].mean().sort_values(ascending=False)
Regalos.head(5)
```

```python
filtro = dataset['gift_type'] == 'Electronics'
df_filtrado = dataset[filtro]
paises = df_filtrado.groupby('country')['country'].count().sort_values(ascending=False)
paises.head()
```

```python
eco = dataset ['price'] < 20
econo = dataset[eco]
economico = econo.groupby(['gift_name'])['price'].sum().sort_values(ascending=True)
economico.head(5)
```

```python
Regalos.plot(kind='bar', title='Regalos por edad mas solicitados')
```

```python
paises.plot(kind='bar', title='Lista de paises que mas regalos solicitan')
```

```python
economico.plot(kind='bar', title='Lista de regalos: Del mas economico al mas caro')
```
